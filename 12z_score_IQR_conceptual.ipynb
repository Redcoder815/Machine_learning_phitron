```
#Step 1 : Generate Normal Data

data  = np.random.normal(loc = 50 , scale =10, size = 1000)

#Step 2

mean = np.mean(data)
std = np.std(data)


#step 3 - convert to z score

z_scores = (data - mean) / std

#step 4

print("original mean", mean)
print("original Std",std)
```

```
#step 5 - visulization

plt.subplot(1,2,1)

sns.histplot(data,bins=30,kde = True)
plt.title("Original Normal distribution")


# Z score distribution

plt.subplot(1,2,2)

sns.histplot(z_scores,bins=30,kde = True )
plt.title("Z socre distribuition")
```

```
data = [12,45,67,68,70,72,73,75,76,80,81,82,84,91,100]

#step 2
data.sort()
print(data)
```

```
#step 3 find median (Q2)

n = len(data)
mid = n // 2

print(mid)

Q2 = data[mid]

print(Q2)
```

```
#lower half, upper half

lower_half = data[:mid]
upper_half = data[mid+1:]

print(lower_half)
print(upper_half)
```

```
# Q1

n1 = len(lower_half)
mid1 = n1//2
Q1 = lower_half[mid1]

print(Q1)
```

```
n2 = len(upper_half)
mid2 = n2//2
Q3 = upper_half[mid2]
print(Q3)
```

```
IQR = Q3 - Q1
print(IQR)
```

```
# lower bound

lower_bound = Q1 - 1.5 * IQR
upper_bound  = Q3 + 1.5 * IQR
```

```
outliers = []

for x in data:
  if x < lower_bound or x> upper_bound:
    outliers.append(x)
```
