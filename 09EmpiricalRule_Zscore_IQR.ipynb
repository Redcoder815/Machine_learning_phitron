1.  According to the Empirical Rule, what percentage of data lies within:

Answer:

±1σ - 68%

±2σ - 95%

±3σ - 99.7%


2. A dataset has a mean of 50 and a standard deviation of 10. According to the Empirical Rule, find the range where approximately 95% of the data lies.

Answer:

mean = 50

standard deviation 10

Approximately 95% of the data falls within 2 standard deviations of the mean (mean ±2σ)

Find the lower and upper bound:

Lower limit = 50 - 2(10) = 50 - 20 = 30

Upper limit = 50 + 2(10) = 70

so range from 30 to 70 has 95% of data

3. A student scored 85 in an exam where the mean score is 70 and the standard deviation is 5. Calculate the Z-score of the student.

Answer:

Student score = 85

mean = 70

standard deviation = 5

so Z-score is:

Z-score = (Student score - mean) / standard deviation

Z-score = (85-70)/5 = 3



4. Given Q1 = 25 and Q3 = 45, calculate the IQR and explain what it represents about the dataset.

Answer:

IQR = Q3-Q1 = 45 - 25 = 20

6. The heights of students are normally distributed with a mean of 160 cm and a standard deviation of 8 cm.

standard deviation 8 cm
