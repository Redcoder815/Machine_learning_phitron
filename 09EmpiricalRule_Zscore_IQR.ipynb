1.  According to the Empirical Rule, what percentage of data lies within:

2. A dataset has a mean of 50 and a standard deviation of 10. According to the Empirical Rule, find the range where approximately 95% of the data lies.

Answer:

mean = 50

standard deviation 10

Approximately 95% of the data falls within 2 standard deviations of the mean (mean ±2σ)

Find the lower and upper bound:

Lower limit = 50 - 2(10) = 50 - 20 = 30

Upper limit = 50 + 2(10) = 70

so range from 30 to 70 has 95% of data

3. A student scored 85 in an exam where the mean score is 70 and the standard deviation is 5. Calculate the Z-score of the student.

Answer:

Student score = 85

mean = 70

standard deviation = 5

so Z-score is:

Z-score = (Student score - mean) / standard deviation

Z-score = (85-70)/5 = 3


A Z-score of 3 means a data point is exactly 3 standard deviations above the mean of its dataset

Key Interpretations

Relative Position: The value is significantly higher than the average or typical data point.

Extremely Rare: In a standard normal distribution, a Z-score of 3 places the value in the 99.87th percentile.

Population Comparison: Only about 0.13% of the entire dataset scores higher than this value.

Outlier Status: In many statistical analyses and machine learning preprocessing steps, any Z-score greater than 3 (or less than -3) is flagged as an outlier.

Real-World Example

If an exam has an average score of 70 with a standard deviation of 5, a student with a Z-score of 3 scored an 85 (70+3x5), meaning they outperformed nearly every other test-taker.

4. Given Q1 = 25 and Q3 = 45, calculate the IQR and explain what it represents about the dataset.

Answer:

IQR = Q3-Q1 = 45 - 25 = 20

An Interquartile Range (IQR) of 20 means that the middle 50% of your data spans a spread of exactly 20 units.

Key Interpretations

Core Spread: It measures the distance between the 25th percentile (Q1) and the 75th percentile (Q3).

Data Packing: The middle half of your dataset is relatively compact or spread out within a 20-unit window (for example, between values of 25 and 45).

Outlier Resistance: This value completely ignores the highest 25% and lowest 25% of the data. Extreme outliers do not affect it.

Real-World Example

If you are analyzing employee salaries and the IQR is 20,000 USD, it tells you that the middle-income half of the company has a salary difference of 20,000 USD between the lower-middle and upper-middle tiers.

8. What is percentile?

Answer:

A percentile is a statistical measure that indicates the percentage of data points in a dataset that fall below a specific value.

It does not represent a score or a percentage correct; instead, it shows your relative standing or rank within a group.

💡 Core Breakdown

Definition: If a value is at the K-th percentile, it means K% of the total data points are equal to or lower than that value.

Scale: Percentiles run from the 1st percentile (lowest) to the 99th percentile (highest).

Median: The 50th percentile represents the exact middle of the dataset (the median).

📝 Real-World Examples
Example 1: Standardized Testing (SAT or GRE)

If you take a test and receive a report stating you scored in the 85th percentile:

Meaning: You scored higher than (or equal to) 85% of all the people who took that test.

Remaining: Only 15% of the test-takers performed better than you.

Example 2: Pediatric Growth Charts

A doctor tells parents that their 2-year-old child's height is in the 10th percentile:

Meaning: The child is taller than or equal to 10% of all 2-year-olds.

Remaining: 90% of children their age are taller than them. (This suggests the child is relatively short for their age group).

🔄 Connection to Quarters (Quartiles)

You can tie percentiles directly back to the IQR (Interquartile Range) concepts you looked at earlier:

25th Percentile (Q1): The value below which 25% of the data lies.

50th Percentile (Median): The value below which 50% of the data lies.

75th Percentile (Q3): The value below which 75% of the data lies.
